```python
import requests
from bs4 import BeautifulSoup
import json
import time
import os
```

# Достаем все ссылки из ЯКласс (запускается 1 раз)

```python
geometry_page = 'https://www.yaklass.ru/p/geometria'
req = requests.get(geometry_page)
src = req.text
soup = BeautifulSoup(src, 'html.parser')
h4 = soup.find_all('h4')

all_lessons_dict = {}
for link in h4:
    lesson_name = link.find('a').text
    lesson_href = 'https://www.yaklass.ru' + link.find('a').get('href')
    all_lessons_dict[lesson_name] = lesson_href
    
# with open('all_lessons_dict.json', 'w', encoding='utf-8') as file:
#     json.dump(all_lessons_dict, file, indent = 4, ensure_ascii = False)
```

```python
with open('all_lessons_dict.json', 'r', encoding='utf-8') as file:
    all_lessons_dict = json.load(file)

all_lessons_dict_items = all_lessons_dict.items()

for lesson_name, lesson_href in all_lessons_dict_items:
    req = requests.get(lesson_href)
    src = req.text
    soup = BeautifulSoup(src, 'html.parser')
    theory_block = soup.find_all('section', class_ = 'TheoryBlock')[0].find_all('a') #находит все ссылки на теорию
    all_lessons_dict[lesson_name] = {}
    for theory in theory_block:
        theory_name = theory.text
        theory_href = 'https://www.yaklass.ru' + theory.get('href')
        print(f'{theory_name} : {theory_href}')
        all_lessons_dict[lesson_name][theory_name] = theory_href
        time.sleep(2)
print(all_lessons_dict)

with open('all_lessons_dict_2.json', 'w', encoding='utf-8') as file:
    json.dump(all_lessons_dict, file, indent = 4, ensure_ascii = False)
```

```python
with open('all_lessons_dict_2.json', 'r', encoding='utf-8') as file:
    all_lessons_dict = json.load(file)

directory = 'YaKlass'
if not os.path.exists(directory):
    os.makedirs(directory)
os.chdir(directory)

all_lessons_dict_items = all_lessons_dict.items()

i = 0
for lesson_name, theory_dict in all_lessons_dict_items:
    for theory_name, theory_href in theory_dict.items():
        req = requests.get(theory_href)
        src = req.text
        soup = BeautifulSoup(src, 'html.parser')
        path = lesson_name.replace(' ','_') + '__' + theory_name.replace(' ','_') + '.html'
        print('Создан файл ', path)
        with open(path, 'w', encoding='utf-8') as file:
            blockbody = soup.find_all(class_ = 'blockbody')
            if len(blockbody) > 1:
                print('Ошибка в ', theory_href)
            file.write(blockbody[0].prettify())
        time.sleep(4)
            
        i += 1
print('Конец')    
```
